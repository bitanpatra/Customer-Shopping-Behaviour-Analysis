```python
import pandas as pd
import numpy as np 
import warnings
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
# check statistics of numerical columns
```

```python
df.isnull().sum()
# checking missing values
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
# filling the nulls with median values
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})

# renaming the columns
```

```python
df.columns.to_list()
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group']= pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)

# first we created a dictionary called frequency_mapping that maps each textual frequency to equivaliant no of days, then we map function to replace text in the column with the corresponding no of days
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
df = df.drop('promo_code_used', axis=1)
```
